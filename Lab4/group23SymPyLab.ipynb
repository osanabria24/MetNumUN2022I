## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = 23 * x1 + 23 * i* x2, y = 10 * y1 + i* y2, z = z1 + i * z2,  x * y * z




Captura de pantalla 2022-04-21 104016.png

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)  
x = 23*x1 + 23*I*x2
y = 10*y1 + I*y2
z = 3*z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))

w=N(1/(pi + I), 20)
print('w=',w)
```
